# Ejercicio 1

### **1. Ecuación diferencial:**
La ecuación original es:

$$
\frac{dy}{dx} = -\frac{x}{y}
$$

Este es el punto de partida. Es una ecuación diferencial que relaciona las derivadas de \( y \) con respecto a \( x \) y se puede resolver usando el método de separación de variables.

---

### **2. Separación de variables:**
El objetivo en este paso es reorganizar la ecuación para que las variables \( y \) y \( x \) queden en lados opuestos. Lo que hice fue multiplicar ambos lados de la ecuación por \( y \, dx \):

$$
y \, dy = -x \, dx
$$

Esto me permite separar las variables $( y ) y ( x )$, de manera que ahora están listas para integrar por separado.

---

### **3. Integración:**
Con las variables separadas, ahora integramos ambos lados:

$$
\int y \, dy = \int -x \, dx
$$

La integral de $( y , dy )$ es $( \frac{y^2}{2} )$, y la integral de $( -x , dx )$ es $( -\frac{x^2}{2} )$, añadiendo una constante $( C )$ de integración:

$$
\frac{y^2}{2} = -\frac{x^2}{2} + C
$$

Este paso es esencial para llegar a una solución general de la ecuación diferencial.

---

### **4. Multiplicación por 2:**
Para simplificar, multiplicamos ambos lados de la ecuación por 2 para eliminar los denominadores:

$$
y^2 = -x^2 + 2C
$$

Ahora la ecuación es más sencilla de manejar.

---

### **5. Aplicación de la condición inicial \( y(4) = 3 \):**
En este paso, aplicamos la condición inicial dada \( y(4) = 3 \). Esto significa que cuando \( x = 4 \), \( y = 3 \). Sustituimos estos valores en la ecuación:

$$
3^2 = -(4^2) + 2C
$$

Esto se convierte en:

$$
9 = -16 + 2C
$$

Resolviendo para \( C \):

$$
9 + 16 = 2C
$$

$$
25 = 2C
$$

$$
C = \frac{25}{2}
$$

Aquí encontramos el valor de la constante de integración \( C \), lo cual nos ayudará a obtener la solución particular de la ecuación.

---

### **6. Sustitución de \( C \) en la solución general:**
Con \( C = \frac{25}{2} \), sustituimos este valor en la solución general que obtuvimos en el paso 4:

$$
y^2 = -x^2 + 25
$$

---

### **7. Solución final:**
Finalmente, tenemos la ecuación:

$$
y^2 = 25 - x^2
$$

Si despejamos \( y \), obtenemos:

$$
y = \pm \sqrt{25 - x^2}
$$

Esta es la solución particular de la ecuación diferencial con la condición inicial \( y(4) = 3 \).

# Ejercicio 2

# Dada la ecuación diferencial:

\begin{equation}
\frac{dy}{dx} = e^{3x + 2y}
\end{equation}

# 1. Separación de variables

Para separar las variables, reescribimos la ecuación utilizando las propiedades de los exponentes:

\begin{equation}
\frac{dy}{dx} = e^{3x} \cdot e^{2y}
\end{equation}


Separando las variables \(x\) y \(y\):


\begin{equation}
e^{-2y} \, dy = e^{3x} \, dx
\end{equation}

---
# 2. Integración de ambos lados


Integramos ambos lados de la ecuación con respecto a sus respectivas variables:


\begin{equation}
\int e^{-2y} \, dy = \int e^{3x} \, dx
\end{equation}


Integral del lado izquierdo
Usamos la sustitución \( u = -2y \), \( du = -2 \, dy \):


\begin{equation}
\int e^{-2y} \, dy = \int e^{u} \left( -\frac{1}{2} \right) \, du = -\frac{1}{2} e^{u} + C = -\frac{1}{2} e^{-2y} + C
\end{equation}


Integral del lado derecho
La integral del lado derecho es directa:


\begin{equation}
\int e^{3x} \, dx = \frac{1}{3} e^{3x} + C
\end{equation}

---
# 3. Igualar ambas integrales


Igualamos las integrales obtenidas:


\begin{equation}
-\frac{1}{2} e^{-2y} = \frac{1}{3} e^{3x} + C
\end{equation}

---
# 4. Despejar la solución general


Multiplicamos ambos lados por \(-2\) para simplificar:


\begin{equation}
e^{-2y} = -\frac{2}{3} e^{3x} - 2C
\end{equation}


Renombramos la constante: \( -2C = C' \):


\begin{equation}
e^{-2y} = -\frac{2}{3} e^{3x} + C'
\end{equation}


Tomamos el logaritmo natural de ambos lados para despejar \(y\):


\begin{equation}
-2y = \ln\left( -\frac{2}{3} e^{3x} + C' \right)
\end{equation}


Finalmente, despejamos \(y\):


\begin{equation}
y = -\frac{1}{2} \ln\left( -\frac{2}{3} e^{3x} + C' \right)
\end{equation}

---
# 5. Solución general


La solución general de la ecuación diferencial es:


\begin{equation}
y = -\frac{1}{2} \ln\left( -\frac{2}{3} e^{3x} + C' \right)
\end{equation}

donde \( C' \) es una constante arbitraria.

# Ejercicio 3

# Dada la ecuación diferencial:

$$
x \frac{dy}{dx} + y = e^x
$$

Se nos pide resolverla usando el **método del factor integrante**. 
## 1. Reescribir la ecuación en forma estándar

La ecuación está en la forma:

$$
\frac{dy}{dx} + \frac{1}{x} y = \frac{e^x}{x}
$$

---
Ahora está en una forma que podemos resolver usando un factor integrante.

## 2. Calcular el factor integrante
El factor integrante se calcula usando la siguiente fórmula:

$$
\mu(x) = e^{\int P(x) \, dx}
$$

Donde $$(P(x) = \frac{1}{x})$$ es el coeficiente que acompaña a $$( y )$$ en la ecuación diferencial.

Entonces:

$$
\mu(x) = e^{\int \frac{1}{x} \, dx} = e^{\ln|x|} = |x|
$$

Podemos asumir $$( \mu(x) = x )$$ ya que trabajamos en valores positivos de $$( x )$$

---
## 3. Multiplicar toda la ecuación por el factor integrante

Multiplicamos ambos lados de la ecuación original por \( x \):

$$
x \frac{dy}{dx} + y = e^x
$$

Esta expresión se convierte en:

$$
\frac{d}{dx}(x y) = e^x
$$

---
## 4. Integrar ambos lados

Ahora integramos ambos lados con respecto a \( x \):

$$
x y = \int e^x \, dx = e^x + C
$$

---
## 5. Despejar \( y \)

Finalmente, despejamos \( y \):

$$
y = \frac{e^x + C}{x}
$$

## Solución final

La solución general de la ecuación diferencial es:

$$
y(x) = \frac{e^x + C}{x}
$$

Donde \( C \) es la constante de integración.
